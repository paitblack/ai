Robot Vacuum Cleaner

-environment matrix  
-1: dirty 0:clean  
-Check if it is dirty --> decision making  
-clean if 1 -> move next one  
-current state : (0,0) 
-move like right - down - left 

```python
class VacuumAgent():

    def __init__(self,environment):
        self.env = environment
        self.state = (0,0)
        self.direction = "right"
    
    def location(self):
        return self.env[self.state[0]][self.state[1]]
    
    def isDirty(self):
        return self.location() == 1
    
    def cleanIT(self):
        self.env[self.state[0]][self.state[1]] = 0
        return self.env

    def move(self,state):
        if self.direction == "right":
          if self.state[1] < len(self.env[0])-1: 
              self.state = (self.state[0],self.state[1]+1)
          else: 
            if self.state[0] < len(self.env)-1: 
              self.state = (self.state[0]+1,self.state[1])
              self.direction = "left"
            else:
              return False
        elif self.direction == "left":
          if self.state[1] > 0: 
              self.state = (self.state[0],self.state[1]-1)
          else: 
            if self.state[0] < len(self.env)-1: 
              self.state = (self.state[0]+1,self.state[1])
              self.direction = "right"
            else:
              return False

        return True
    
    def act(self):

        while True:
          print(f"Agent at: {self.state}")
          if self.isDirty():
              print(f"Current position: Dirty")
              print(f"Agent action: Clean")
              self.cleanIT()
          else:
              print(f"Current position: Clean")

          print(f"Agent action: Move")
          if self.move(self.state):
             print(f"Agent moved to: {self.state}")
          else:
            break

        print(f"No more positions to explore!")
        print(f"Agent stoped at: {self.state}")
        return self.env

environment = [[1, 0], [0, 1]]
print(f"Initial Environment: {environment}\n")

agent = VacuumAgent(environment)

agent.act()

print(f"\nResulting Environment: {environment}")            
        

```
